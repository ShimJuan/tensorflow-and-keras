```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 공부한시간, 출석일수
xdata = [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]]
ydata = [[0],[0],[0],[1],[1],[1]] # 0 : 불합격 / 1 : 합격
```

```python
x = tf.placeholder(tf.float32, shape=[None,2])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([2,1]))
b = tf.Variable(tf.random_uniform([1]))
```

```python
z = tf.matmul(x,w)+b
hx = tf.nn.sigmoid(z)
```

```python
cost_i = tf.nn.sigmoid_cross_entropy_with_logits(logits=z, logits=y)
# y*(-tf.log(hx))+(1-y)*(-tf.log(1-hx))
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.AdamOptimizer( 0.1 )
train = optimizer.minimize( cost )
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(1000):
    sess.run(train,{x:xdata})
    print(i,sess.run(cost, {x:xdata}))
```
