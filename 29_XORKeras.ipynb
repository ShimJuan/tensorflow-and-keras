```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

from sklearn import preprocessing
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
xdata = np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32)
ydata = np.array([[0],[1],[1],[0]],dtype=np.float32)
```

```python
# IO = Dense()
# model = Sequential([IO])
model = Sequential()
model.add(Dense(units=10, input_dim=2, activation='sigmoid'))
model.add(Dense(units=1, input_dim=10, activation='sigmoid')) # 두번째 input dim은 두번째부턴 빼도 된다. 케라스에서 자동으로 잡아주기 때문에
model.add(Dense(units=1, input_dim=10, activation='sigmoid')) # 첫째줄의 2와 셋째줄의 1만 맞추면 된다
# dense 갯수를 deep하게  
model.compile(loss='binary_crossentropy',
              optimizer=Adam(0.1),metrics=['accuracy'])
h = model.fit(xdata,ydata, epochs=2000)
```

```python
model.predict_classes(xdata)
```

```python
h.history['acc'][-1]
```

## 학습 정리 
#### sigmoid / softmax 
- sigmoid는 기다 아니다 ( yes or no)
- softmax는 어떠한 값에 대해 예측을 하고싶을때 사용한다.
- 결론: 똑같은 데이터셋이더라도 의도에 따라 다른 함수를 사용한다.
