```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 헤더가 없기때문에 칼럼을 비지정한후 칼럼명을 따로 설정한다
df = pd.read_csv('data-01.csv',header=None)
df.columns=['q1','q2','mid','score']
df
```

```python
# 데이터프레임의 형태를 유지하려면 []를 한번 더 씌워준다
xdata = df[['q1','q2','mid']]
ydata = df[['score']]
```

```python
x = tf.placeholder(tf.float32, shape=[None,3])
y = tf.constant(ydata,tf.float32)
```

```python
# [3,1] #[피쳐갯수, 라벨갯수] : matmul을 이용시에
w = tf.Variable(tf.random_uniform([3,1])) # 3행 1열로 출력되어야 하기 때문에 미리 정해준다
b = tf.Variable(tf.random_uniform([1]))
```

```python
hx = tf.matmul(x,w)+b # n행3열, 3행1열
cost = tf.reduce_mean(tf.square(hx-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.00001) #learning rate
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(10000):
    sess.run(train, {x:xdata})
    print(i,sess.run(cost, {x:xdata}))
```
