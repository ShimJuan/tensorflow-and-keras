```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
```

```python
iris = datasets.load_iris()
```

```python
xdata = iris['data']
ydata = to_categorical(iris['target'],3)
```

```python
a,b,c,d = [np.array([1,2]),np.array([3,4]),
           np.array([5,6]),np.array([7,8])]
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(xdata,ydata,test_size=0.3) 
# 30프로의 사이즈로 학습시킨다.
```

```python
x = tf.placeholder(tf.float32,[None,4])
y = y_train
w = tf.Variable(tf.random_uniform([4,3]))#[f갯수,분류3]
b = tf.Variable(tf.random_uniform([3]))
```

```python
z = tf.matmul(x,w) + b
hx = tf.nn.softmax(z) # 예측값 확인용도 , 밑의 함수가 있을경우 굳이 사용할필요없다.
cost_i = tf.nn.softmax_cross_entropy_with_logits_v2(logits=z,labels=y) # label : 실제값
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1) #learning rate
train = optimizer.minimize(cost)
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(3000):
    sess.run(train,{x:x_train}) # train 데이터셋 이용 
    c = sess.run(cost,{x:x_train})
    print(i,c)
```

```python
# 정확도를 구한다 
t = sess.run(hx,{x:x_test})
```
