### XOR problem 
- pdf 129~

0 or 1이냐의 분류문제를 해결하는것은 logistic

hx = sigmoid(z)를 통과
이미지 파일 등을 이런식으로 학습한다.

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
```

```python
xdata = np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32)
ydata = np.array([[0],[1],[1],[0]],dtype=np.float32)
```

```python
x = tf.placeholder(tf.float32,[None,2])
y = tf.constant(ydata, tf.float32)
```

```python
w = tf.Variable(tf.random_uniform([2,1]))
b = tf.Variable(tf.random_uniform([1]))
```

```python
z = tf.matmul(x,w)+b
hx = tf.nn.sigmoid(z)
```

```python
cost_i=tf.nn.sigmoid_cross_entropy_with_logits(logits=z,labels=y)
cost = tf.reduce_mean( cost_i )
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1) #learning rate
train = optimizer.minimize(cost)
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(3000):
    sess.run(train,{x:xdata}) #벼림 연산 # w = w.assign(w-0.1*g)
    c = sess.run(cost,{x:xdata})
    print(i,c)
```

```python
h = sess.run(hx,{x:xdata})>0.5
t = tf.cast(h>0.5,tf.float32)
sess.run(t)
```

###### 결론:  학습을 아무리 시켜도 0,1,1,0은 나오지 않는다.
