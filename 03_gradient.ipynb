```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [1,2,3]
```

```python
def cost(x,y,w): 
    c = 0
    for i in np.arange(len(x)):
        hx = w*x[i]
        c = c+(hx-y[i])*x[i] 
    return c/len(x)
```

```python
def gradient(x,y,w): #기울기
    c = 0
    for i in np.arange(len(x)):
        hx = w*x[i]
        c = c+(hx-y[i])*x[i] # 편미분
    return c/len(x)
```

```python
def show_gradient():
    x = [1,2,3]
    y = [1,2,3]
    w = 10
    for i in range(200):
        c = cost(x,y,w)
        g = gradient(x,y,w)
        w = w-0.1*g
        print(i,c)
    print('w:',w)
show_gradient()
```
